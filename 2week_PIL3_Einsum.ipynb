rectangle 그리고 그 위에 text 넣기(yolo 등에서 box 만들때)

```python
jang = Image.open('1.jpg')
font = ImageFont.truetype('malgun.ttf',100)
draw = ImageDraw.Draw(jang)
text = '원영'
text_width, text_height = font.getsize(text)
x, y = 650, 350
```

```python
draw.rectangle(((x,y - text_height),(x + text_width, y)),fill='pink')
draw.text((x,y - text_height),text,font=font,fill=(255,255,255))
jang.resize((200,300))
```

**matplotlib과 mask로 그림 그리기**

```python
xx,yy = np.ogrid[:100,:100]
mask = yy - 2*xx + 3 == 0
condinator = np.ones((100,100))
condinator[mask] = 255  # 흰색 표시
```

```python
import matplotlib.pyplot as plt
plt.xticks([1,2,3,])
plt.imshow(condinator)
```

```python
mask = (yy-20)**2 + (xx-20)**2 - 40 == 0
condinator = np.ones((100,100))
condinator[mask] = 255
plt.imshow(condinator)
```

```python
mask = (yy-50)**2 + (xx-50)**2 - 40 <= 0
condinator = np.ones((100,100))
condinator[mask] = 255
plt.imshow(condinator)
```

## Einsum

- **규칙에 따라서 간결하게 벡터와 합연산을 함께 표현**
- 딥러닝에서 vectorization 이용해서 핵심임
- 벡터 행렬 곱의 합을 한 줄로 표현

</br>

- -> : 합(뒤에 값없으면 생략가능)
- 인자 개수만큼 string으로 들어가고 ,으로 구분

```python
a = np.array([0,1,2,3])
aa = np.arange(6).reshape(2,3)
aa
```

```python
np.einsum("i,i",a,a) #내적
```

```python
np.inner(a,a) # np.dot(a,a)
```

```python
np.einsum("ij",aa)  # 차원만큼 개수표현
```

```python
np.einsum("ab,ba",aa,aa.T)
```

```python
np.einsum("ba,ab",aa,aa.T)
```

```python
np.matmul(aa,aa.T)
```

```python
np.einsum('ab->ba',aa) # 전치
```

```python
aa = np.arange(4).reshape(2,2)
bb = np.arange(1,5).reshape(2,2)
print(aa)
print(bb)
np.einsum('ab,ap->bp',aa,bb)
```

```python
np.einsum('ij->i',bb) # col 합
```

```python
np.einsum('ab,bc->ac',aa,bb) # 행렬곱
```

```python
np.einsum('xx->x',aa) # 대각선
# np.einsum('xx->',aa) -> 3
```

```python
np.einsum('x,x->',a,a)
```

```python
np.einsum('ab,bc->a',aa,bb) # col합
np.einsum('ab,bc->c',aa,bb) # row합
```

```python
np.einsum('ab,bc->abc',aa,bb)
```

```python
np.einsum('ab,bc->b',aa,bb)
# element-wise?
```
